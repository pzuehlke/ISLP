```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

__12 (a):__ The following formula for the cofficient estimate $ \hat{\beta} $
for the regression line of $ y $ onto $ x $ was established in Exercise 11:
$$
    \hat{\beta} = \frac{\mathbf{x} \cdot \mathbf{y}}{\mathbf{x} \cdot \mathbf{x}}
                = \frac{\mathbf{x} \cdot \mathbf{y}}{\Vert \mathbf x \Vert^2}
$$
By symmetry, for the opposite regression of $ x $ on $ y $, the
associated coefficient $ \gamma $ is estimated to be
$$
    \hat{\gamma} = \frac{\mathbf{y} \cdot \mathbf{x}}{\mathbf{y} \cdot \mathbf{y}}
                 = \frac{\mathbf{y} \cdot \mathbf{x}}{\Vert \mathbf y \Vert^2}

$$
Thus, the two coincide if and only if
$$
\Vert \mathbf x \Vert^2 = \sum_{i=1}^n x_i^2 =  \sum_{i=1}^n y_i^2 = \Vert \mathbf y \Vert^2 \,.
$$
If both samples have average equal to zero, then this condition is equivalent to
saying that the respective sample standard deviations of $ x $ and $ y $
coincide.

__12 (b):__ We only need to generate samples $ \mathbf{x} $ and $ \mathbf{y} $ such that
$ \Vert \mathbf{x} \Vert = \Vert \mathbf{y} \Vert $. For example:

```python
rng = np.random.default_rng(1)
x = rng.normal(size=100)
y = rng.normal(size=100)

# Multiply y by a factor so that the result has the same norm as x:
y *= np.linalg.norm(x) / np.linalg.norm(y)
```

```python
model_y_on_x = sm.OLS(y, x).fit()
model_y_on_x.summary()
```

```python
model_x_on_y = sm.OLS(x, y).fit()
model_x_on_y.summary()
```

__12 (c):__ Pretty much any generated example will work. Let's take an example where both $ y $ and $ x $ are taken from a population having mean zero.

```python
rng = np.random.default_rng(1)
x = rng.normal(size=100)
y = 2 * x + rng.normal(size=100)

fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(x, y, s=30, color="black");
```

```python
model_y_on_x = sm.OLS(y, x).fit()
model_y_on_x.summary()
```

```python
model_x_on_y = sm.OLS(x, y).fit()
model_x_on_y.summary()
```
