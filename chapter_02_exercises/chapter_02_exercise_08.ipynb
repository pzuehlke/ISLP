__8 (h):__ We begin by computing the acceptance rate of the colleges and
computing its correlation with some of the other variables.

```python
# Create a new column for the acceptance rate:
college["Acceptance.Rate"] = college["Accept"] / college["Apps"] * 100

print("\nCorrelations with acceptance rate:")
grad_correlations = college[["Acceptance.Rate", "Apps", "Enroll", "Top10perc", "perc.alumni",
                             "Personal", "PhD", "Grad.Rate"]].corr()["Acceptance.Rate"].sort_values(ascending=False)
print(grad_correlations)
```

Most of these results are not surprising, at least qualitatively. For example, we see that the higher
the acceptance rate (that is, the easier it is to get into a given college), the smaller the percentage
of new students who were in the top $ 10\% $ of their high-school class. It is interesting that the
estimated personal spending seems quite uncorrelated to the acceptance rate.

We now consider some differences among private and public colleges:

```python
print("\nSummary statistics by college type:", end="\n\n")
metrics1 = ["Apps", "Acceptance.Rate", "Top10perc"]
metrics2 = ["Outstate", "Expend", "Grad.Rate"]
print(college.groupby("Private")[metrics1].agg(["mean", "median"]).round(2), end="\n\n")
print(college.groupby("Private")[metrics2].agg(["mean", "median"]).round(2))
```
