__15 (d):__ 

```python
def Power3(x, a):
    return x**a


Power3(8, 17)
```

__15 (e):__ 

```python
fig, ax = plt.subplots(figsize=(8, 5))
xs = np.arange(1, 11)
ys = Power3(xs, 2)
ax.scatter(xs, ys, color="black", s=20) 

ax.set_xlabel("$ x $")
ax.set_ylabel("$ y $")
ax.set_title("Squares of integers from $ 1 $ to $ 10 $")
ax.grid(True)
ax.set_xscale("log", base=2)
ax.set_yscale("log", base=2)
```

__15 (f):__ 

```python
def PlotPower(xs, a):
    fig, ax = plt.subplots(figsize=(8, 5))
    ys = Power3(xs, a)
    ax.scatter(xs, ys, color="black", s=20) 

    ax.set_xlabel("$ x $")
    ax.set_ylabel("$ y $")
    ax.set_title("Powers that be")
    ax.grid(True)


PlotPower(np.arange(1, 11), 3)
```
